# Train?

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
```

```python
def feature_engineering (df):
    language_encode = pd.get_dummies(df['Language'])
    df = df.drop(columns=['Mature','Partnered'])
    df['isSpanish'] = df['Language'] == 'Spanish'
    df['Retention ratio'] = df['Average viewers'] / df['Peak viewers']
    df['Watch time ratio'] = df['Watch time(Minutes)'] / df['Stream time(minutes)']
    df['Viewer Ratio'] = df['Average viewers'] / df['Followers']
    df = df.drop(columns=['Language'])
#     df_result = pd.concat([df_result,language_encode], sort=False, axis = 1)
    return df
```

```python
def splitXY (df):
    Y_train = df['Followers gained']
    X_train = df.drop(columns=['Followers gained'])
    return X_train,Y_train
```

```python
df_1 = feature_engineering (df)
```

```python
X_train, Y_train = splitXY(df_1)
```

```python
print('Train X shape: ',np.shape(X_train))
print('Train Y shape: ',np.shape(Y_train))
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor(n_estimators=200)
```

```python
forest_reg.fit(X_train[:700], Y_train[:700])
prediction = forest_reg.predict(X_train[700:])
mse = mean_squared_error(Y_train[700:], prediction)
rmse = np.sqrt(mse)
rmse
```

# Submission 만들기

```python
df_test_1 = feature_engineering (df_test)
df_test_1 = scaler.transform(df_test_1)
```

```python
forest_reg = RandomForestRegressor(n_estimators=100)
forest_reg.fit(X_train, Y_train)
prediction_test = forest_reg.predict(df_test_1)
```

```python
df_test['Followers gained'] = prediction_test
df_pred = pd.DataFrame(df_test['Followers gained'])
df_pred.to_csv('submission_minseok.csv')
```
